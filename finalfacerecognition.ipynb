```
import face_recognition as fr
```

```
image=fr.load_image_file("/content/drive/My Drive/image.jpg")
```

```
from google.colab.patches import cv2_imshow
```

```
cv2_imshow(image)
```

```
locs=fr.face_locations(image)
```

```
print(len(locs),"faces detected")
```

```
print(locs)
```

```
import cv2
```

```
for face in locs:
  cv2.rectangle(image,(face[3],face[0]),(face[1],face[2]),(0,255,0),2)
  
```

```
cv2_imshow(image)
```
